```python
import requests
from bs4 import BeautifulSoup as bs
import pickle
```

```python
def convert_url_to_transcript(url):
    page = requests.get(url).text
    soup = bs(page, "lxml")
    text = [p.text for p in soup.find(class_="post-content").find_all('p')]
    print(url)
    return text
```

```python
# Comedian names
comedians = ['louis', 'dave', 'ricky', 'bert', 'bill', 'jim', 'john', 'hasan', 'ali', 'anthony', 'mike', 'joe', 'russell', 'nikki']
```

```python
#URLs of the transcripts
urls = ['http://scrapsfromtheloft.com/2017/05/06/louis-ck-oh-my-god-full-transcript/',
        'https://scrapsfromtheloft.com/2019/08/26/dave-chappelle-sticks-stones-transcript/',
        'http://scrapsfromtheloft.com/2018/03/15/ricky-gervais-humanity-transcript/',
        'https://scrapsfromtheloft.com/2020/03/21/bert-kreischer-hey-big-boy-transcript/',
        'http://scrapsfromtheloft.com/2017/05/24/bill-burr-im-sorry-feel-way-2014-full-transcript/',
        'http://scrapsfromtheloft.com/2017/04/21/jim-jefferies-bare-2014-full-transcript/',
        'https://scrapsfromtheloft.com/2018/05/05/john-mulaney-kid-gorgeous-at-radio-city-full-transcript/',
        'http://scrapsfromtheloft.com/2017/10/21/hasan-minhaj-homecoming-king-2017-full-transcript/',
        'http://scrapsfromtheloft.com/2017/09/19/ali-wong-baby-cobra-2016-full-transcript/',
        'http://scrapsfromtheloft.com/2017/08/03/anthony-jeselnik-thoughts-prayers-2015-full-transcript/',
        'http://scrapsfromtheloft.com/2018/03/03/mike-birbiglia-my-girlfriends-boyfriend-2013-full-transcript/',
        'http://scrapsfromtheloft.com/2017/08/19/joe-rogan-triggered-2016-full-transcript/',
       'https://scrapsfromtheloft.com/2018/06/29/russell-peters-almost-famous-2016-full-transcript/',
       'https://scrapsfromtheloft.com/2019/10/04/nikki-glaser-bangin-transcript/']
```

```python
# Request transcripts form the URLs (takes a few minutes to run)
transcripts = [convert_url_to_transcript(u) for u in urls]
```

```python
# Pickle files for later use

# Make a new directory to hold the text files
!mkdir transcripts

for i, c in enumerate(comedians):
    with open("transcripts/" + c + ".txt", "wb") as file:
        pickle.dump(transcripts[i], file)
```

```python
#Load the pickled files so that we dont have to scrape the data again

data = {}

for i, c in enumerate(comedians):
    with open("transcripts/" + c + ".txt", "rb") as file:
        data[c] = pickle.load(file)
```

```python
# the value in the dictionary contains multiple chunks (paragraphs) so make it all one value by removing the breaks
def combine_text(input_value):
    '''Take the list of text and combine them into one large text'''
    combined_text = ' '.join(input_value)
    return combined_text
```

```python
# Run the combination function on the data dictionary
data_combined = {key: [combine_text(value)] for (key, value) in data.items()}
```

```python
#data_combined
```

```python
# Move the combined data into a dataframe
import pandas as pd
pd.set_option('max_colwidth', 150)

data_df = pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns = ['transcript']
data_df = data_df.sort_index()
data_df
```

```python
#data_df.transcript.loc['ali']
```

```python
import re
import string

#Regular expression data cleanup
def clean_text_round1(text):
    # Make text lowercase
    text = text.lower()
    # Remove everything in square braces
    text = re.sub('\[.*?\]', '', text)
    # Remove all the puncuation marks
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    # Remove words containing numbers
    text = re.sub('\w*\d\w*', '', text)
    return text

round1_cleanup = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data_df.transcript.apply(round1_cleanup))
data_clean
```

```python
#Second round of cleaning
def clean_text_round2(text):
    #get rid of the quotes
    text = re.sub('[''""…]', '', text)
    #get rid of the /n symbol
    text = re.sub('/n', '', text)
    return text

round2_cleanup = lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.transcript.apply(round2_cleanup))
data_clean
```

```python
#Creating the corpus - which is a collection of text
full_names = ['Ali Wong', 'Anthony Jeselnik', 'Bill Burr', 'Bo Burnham', 'Dave Chappelle', 'Hasan Minhaj',
              'Jim Jefferies', 'Joe Rogan', 'John Mulaney', 'Louis C.K.', 'Mike Birbiglia', 'Ricky Gervais']
data_df['full_names'] = full_names
data_df
```

```python
#Pickle the DF 
data_df.to_pickle("corpus.pkl")
```

```python
# Creating the document term matrix - which is the matrix of the word count
from sklearn.feature_extraction.text import CountVectorizer

# Using the count vectorizer sop word list to exclude the stop words from our dataset
cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(data_clean.transcript)

data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```

```python
#Pickle the Document Term Matrix
data_dtm.to_pickle("dtm.pkl")
```

```python
#pickle the cleaned data (before we put it in document-term matrix format) and the CountVectorizer object
data_clean.to_pickle('data_clean.pkl')
pickle.dump(cv, open('cv.pkl', 'wb'))
```
